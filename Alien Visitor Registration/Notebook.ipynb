# Duplicate Records Detection

```python
data_cleaned = data
```

```python
data_cleaned.duplicated().sum()
```

```python
data_cleaned = data_cleaned.drop_duplicates()
data_cleaned.duplicated().sum()
```

We notice that sum of dublicated records becames 0

# Planet Names Standardization

```python
column = data_cleaned["Planet"]
column
```

```python
data_cleaned["Planet"] = column.str.strip().str.lower()
data_cleaned["Planet"]
```

Here the text standardization was important because there was a messy data (different format) 

# Missing Values Handling

```python
age = data_cleaned["Age"]
food = data_cleaned["Favorite_Food"]
mean = age.mean()
mode = food.mode()[0]
print(f"mean age: {mean}, median age: {age.median()}")
print(f"most frequent food value: {mode}")
```

```python
data_cleaned["Age"] = age.fillna(mean)
data_cleaned["Favorite_Food"] = food.fillna("Unknown")
data_cleaned.isnull().sum()
```

# Planet Distribution Analysis

```python
data_cleaned["Planet"].value_counts()
```

Mars has the highest number of visitors

# Food Preference Analysis

```python
data_cleaned["Favorite_Food"].value_counts()
```

The most popular food among visitors is "Pizza"

# Age Group Creation

```python
data_cleaned["Age_Group"] = pd.cut(data_cleaned["Age"], bins = 3, labels = ["Young", "Adult", "Senior"])
data_cleaned["Age_Group"].value_counts()
```

# Insight Generation

```python
print(f"""
average visitor age: {mean}
most common planet: {data_cleaned["Planet"].mode()[0]}
most popular food: {mode}
most frequent age group: {data_cleaned["Age_Group"].mode()[0]}
""")
```

At the end, we notice that data preperation improved the dataset and made it more accurate and not messy, and these cleaning steps were important before analysis because they prevent the inaccurate analysis

```python
data_cleaned.to_csv("data/alien_visitors_cleaned.csv", index = False)
```
