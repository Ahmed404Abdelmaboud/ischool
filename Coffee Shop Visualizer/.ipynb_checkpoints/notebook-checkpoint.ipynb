# Library Initialization

```python
import matplotlib.pyplot as plt
```

# Data

```python
days = list(range(1, 16))
daily_sales = [120, 135, 140, 138, 150, 155, 160, 158, 170, 175, 180, 190, 195, 200, 210]
drinks = ['Espresso', 'Latte', 'Cappuccino', 'Americano', 'Mocha']
cups_sold = [45, 60, 50, 35, 40]
prices = [2.5, 3.0, 3.5, 4.0, 4.5, 5.0]
sales_at_price = [200, 180, 160, 140, 120, 100]
sales_distribution = [120, 125, 130, 135, 140, 145, 150, 155, 160, 165, 170, 175, 180, 185, 190, 195, 200, 205, 210, 215, 220, 130, 140, 150, 160, 170, 180, 190, 200, 210]
drink_revenue_percent = [35, 30, 20, 10, 5]
```

# Line Plot – Sales Change Over Time

```python
plt.plot(days, daily_sales)
plt.title("Coffee Shop Daily Sales Trend")
plt.xlabel("Day")
plt.ylabel("Sales ($)");
```

# Bar Chart – Drink Popularity

```python
plt.bar(drinks, cups_sold)
plt.title("Number of Cups Sold by Drink Type")
plt.xlabel("Drink")
plt.ylabel("Cups Sold");
```

# Scatter Plot – Price vs Sales Relationship

```python
plt.scatter(prices, sales_at_price)
plt.title("Effect of Price on Sales")
plt.xlabel("Price ($)")
plt.ylabel("Sales (units)")
plt.grid();
```

#### The scatter plot shows a "negative correlation"

# Histogram – Sales Distribution

```python
plt.hist(sales_distribution, bins = 5, edgecolor = "black")
plt.title("Distribution of Daily Sales")
plt.xlabel("Sales ($)")
plt.ylabel("Frequency");
```

#### Most sales concentrated in range 200 - 220

# Pie Chart – Revenue Percentage by Drink

```python
plt.pie(drink_revenue_percent, labels = drinks, autopct = "%1.1f%%");
```

# Final Reflection

- Americano is the least cup sold, so we must improve it or decrease its production
- There is a negative correlation between price and sales, so we must decrease the prices
- Mocha and Americano have the least revenue so, we must improve them or decrease thier production
