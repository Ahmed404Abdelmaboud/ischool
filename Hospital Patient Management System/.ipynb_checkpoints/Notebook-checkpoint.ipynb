# Setup & Data Loading

```python
import pandas as pd

patients = pd.read_csv("data/patients.csv")
appointments = pd.read_csv("data/appointments.csv")
doctors = pd.read_csv("data/doctors.csv")

patients.to_sql("patients", "sqlite:///data/hospital.db", if_exists = "replace", index = False)
appointments.to_sql("appointments", "sqlite:///data/hospital.db", if_exists = "replace", index = False)
doctors.to_sql("doctors", "sqlite:///data/hospital.db", if_exists = "replace", index = False)

%reload_ext sql
%config SqlMagic.style = "_DEPRECATED_DEFAULT"
%sql sqlite:///data/hospital.db
```

# Explore the Data

## Task 1: Show all patients

```python
%%sql
SELECT * FROM patients
```

## Task 2: Show all doctors

```python
%%sql
SELECT * FROM doctors
```

## Task 3: Show all appointments

```python
%%sql
SELECT * FROM appointments
```

# Basic Analysis (SELECT + WHERE)

## Task 4: Find patients from Cairo

```python
%%sql
SELECT * FROM patients WHERE City = "Cairo"
```

## Task 5: Find doctors with specialization in Cardiology

```python
%%sql
SELECT * FROM doctors WHERE Specialization = "Cardiology"
```

## Task 6: Find appointments that were Cancelled

```python
%%sql
SELECT * FROM appointments WHERE Status = "Cancelled"
```

# Sorting Data

## Task 7: Sort patients by age (oldest first)

```python
%%sql
SELECT * FROM patients ORDER BY Age DESC
```

## Task 8: Sort doctors by experience (most experienced first)

```python
%%sql
SELECT * FROM doctors ORDER BY Experience DESC
```

# LIMIT + OFFSET

## Task 9: Show first 3 patients

```python
%%sql
SELECT * FROM patients LIMIT 3
```

## Task 10: Skip first 3 appointments and show next 3

```python
%%sql
SELECT * FROM appointments LIMIT 3 OFFSET 3
```

# Combined Logic

## Task 11: Find patients from Cairo sorted by age

```python
%%sql
SELECT * FROM patients WHERE City = "Cairo"
ORDER BY Age
```

## Task 12: Find most recent appointments (sort by date descending)

```python
%%sql
SELECT * FROM appointments ORDER BY AppointmentDate DESC
```

## Task 13: Show top 2 patients with the highest age

```python
%%sql
SELECT * FROM patients ORDER BY Age DESC LIMIT 2
```
