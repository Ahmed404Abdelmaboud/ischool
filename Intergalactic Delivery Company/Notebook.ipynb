# Label Encoding – Performance Level

```python
df["Performance_Encoded"] = df["Performance_Level"].map({"Low": 0, "Medium": 1, "High": 2})
df["Performance_Encoded"]
```

# One-Hot Encoding

```python
df = pd.get_dummies(df, columns = ["Planet", "Department"])
df
```

# Normalization – Monthly Deliveries: 

```python
column = df["Monthly_Deliveries"]
df["Deliveries_Normalized"] = (column - column.min()) / (column.max() - column.min())
print(df["Deliveries_Normalized"].min(), df["Deliveries_Normalized"].max())
```

# Standardization – Monthly Deliveries

```python
column = df["Monthly_Deliveries"]
df["Deliveries_Standardized"] = (column - column.mean()) / column.std()
print(df["Deliveries_Standardized"].mean(), df["Deliveries_Standardized"].std())
```

# Comparison Analysis

- Label Encoding: Converts categories into numeric values. Best for ordinal data.

- One-Hot Encoding: Converts each category into a separate binary column. Best for nominal data where no order exists.

- Normalization: Scales values into a fixed range (usually 0–1). Useful for machine learning.

- Standardization: it makes mean = 0 and standard deviation = 1. Useful for machine learning.
