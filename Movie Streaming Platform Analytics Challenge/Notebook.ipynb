# Library Initialization

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Dataset Exploration

```python
df = pd.read_csv("data/movie_streaming_users.csv")
df.head(5)
```

```python
df.info()
```

```python
df.shape
```

# Bar Chart Analysis

```python
sns.countplot(df["Favorite_Genre"]);
```

- The most popular genre => Sci-Fi
- The least popular genre => Romance

# Pie Chart Analysis

```python
count = df["Subscription_Type"].value_counts()
plt.pie(count, labels = count.index, autopct = "%1.1f%%");
```

- The largest subscription group => Basic
- The Smallest subscription group => Premium

# Histogram Analysis

```python
sns.histplot(df["Watch_Hours_Per_Month"]);
```

- Range 35-40 is most frequently
- The distibution is spread out

# Box Plot Analysis

```python
sns.boxplot(df["Monthly_Spending"]);
```

- There are 5 potential outliers
- The spending is relatively consistent among users

# Scatter Plot Analysis

```python
sns.scatterplot(data = df, x = "Watch_Hours_Per_Month", y = "Movies_Watched")
sns.regplot(data = df, x = "Watch_Hours_Per_Month", y = "Movies_Watched", color = "green", scatter = False);
```

- There is a positive correlation
- The scatter plot suggests that the usrs who spend long time their consumption increases

# Insight Generation

- The most popular genre => Sci-Fi
- The most common subscription type => Basic
- There is 5 spending outliers
- There is a positive relationship between watch hours and movies watched
- Scatter plot provided the most useful insight
- Selecting the correct chart is important to achieve useful insights
