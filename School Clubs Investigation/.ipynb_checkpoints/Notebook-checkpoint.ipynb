```python
%%sql
CREATE TABLE Students (
    StudentID INT PRIMARY KEY,
    StudentName VARCHAR(50),
    Grade INT,
    City VARCHAR(50)
);
```

```python
%%sql
INSERT INTO Students
VALUES
(1,'Ali',10,'Cairo'),
(2,'Sara',9,'Giza'),
(3,'Omar',10,'Cairo'),
(4,'Noor',8,'Alexandria'),
(5,'Youssef',9,NULL),
(6,'Lina',10,'Cairo');
```

```python
%%sql
CREATE TABLE Clubs (
    ClubID INT PRIMARY KEY,
    ClubName VARCHAR(50),
    Instructor VARCHAR(50)
);
```

```python
%%sql
INSERT INTO Clubs
VALUES
(1,'Robotics','Mr. Ahmed'),
(2,'Art','Ms. Salma'),
(3,'Coding','Mr. Ahmed'),
(4,'Sports','Mr. Kareem');
```

```python
%%sql
CREATE TABLE Memberships (
    MembershipID INT PRIMARY KEY,
    StudentID INT,
    ClubID INT
);
```

```python
%%sql
SELECT * FROM Students WHERE Grade > 9 AND City = "Cairo"
```

```python
%%sql
SELECT * FROM Students WHERE Grade > 9 OR City = "Cairo"
```

```python
%%sql
SELECT * FROM Students WHERE NOT City = "Alexandria"
```

```python
%%sql
SELECT * FROM Students WHERE StudentName LIKE "%r"
```

```python
%%sql
SELECT * FROM Students WHERE StudentName LIKE "%oo%"
```

```python
%%sql
SELECT * FROM Clubs WHERE ClubName LIKE "%s"
```

```python
%%sql
SELECT * FROM Students Where StudentID IS Null OR StudentName IS Null OR Grade IS Null OR City IS Null
```

```python
%%sql
SELECT COUNT(*) FROM Students
```

```python
%%sql
SELECT AVG(Grade) FROM Students
```

```python
%%sql
SELECT MAX(Grade), MIN(Grade) FROM Students
```

```python
%%sql
SELECT COUNT(*) FROM Memberships
```

```python
%%sql
SELECT Grade, COUNT(*) AS Students FROM Students GROUP BY Grade
```

```python
%%sql
SELECT City, COUNT(*) AS Students FROM Students GROUP BY City HAVING Students > 1
```

```python
%%sql
SELECT c.ClubName, COUNT(*) AS Students FROM Memberships m JOIN Clubs c ON c.ClubID = m.ClubID GROUP BY c.ClubID HAVING Students > 1 ORDER BY Students DESC
```

```python
%%sql
SELECT c.Instructor, COUNT(*) AS Students FROM Memberships m JOIN Clubs c ON c.ClubID = m.ClubID GROUP BY c.Instructor HAVING Students > 1 ORDER BY Students DESC
```

```python
%%sql
SELECT s.StudentName, c.ClubName FROM Students s JOIN Memberships m ON m.StudentID = s.StudentID JOIN Clubs c ON c.ClubID = m.ClubID
```
