## Data Cleaning Decisions

```python
data["Passengers"] = data["Passengers"].fillna(data["Passengers"].mean())
data["Delay_Minutes"] = data["Delay_Minutes"].fillna(data["Delay_Minutes"].median())
data.isnull().sum()
```

```python
data = data.drop_duplicates()
data.duplicated().sum()
```

```python
data["Transport_Type"] = data["Transport_Type"].str.strip().str.capitalize()
data["Transport_Type"].unique()
```

```python
data["Day_Type"] = data["Day_Type"].str.strip().str.capitalize()
data["Day_Type"].unique()
```
