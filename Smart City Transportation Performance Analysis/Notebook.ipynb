```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Dataset Review

```python
data =  pd.read_csv("data/Cleaned_Data.csv")
data.head()
```

# Descriptive Analysis

```python
data.describe()
```

# Station Analysis

```python
traffic = data.groupby("Station")["Delay_Minutes"].sum().sort_values(ascending = False).reset_index()
traffic
```

```python
bars = sns.barplot(data = traffic, x = "Station", y = "Delay_Minutes")
for bar in bars.patches:
    height = bar.get_height()
    plt.annotate(f"{height}", xy = (bar.get_x() + bar.get_width() / 2, height + 0.5), ha = "center");
```

# Transport Type Analysis

```python
transport = data["Transport_Type"].value_counts().reset_index()
transport
```

```python
plt.pie(transport["count"] / transport["count"].sum() * 100, labels = transport["Transport_Type"], autopct = "%1.1f%%");
```
