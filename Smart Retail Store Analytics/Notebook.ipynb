```python
import pandas as pd

customers = pd.read_csv("data/customers.csv")
orders = pd.read_csv("data/orders.csv")
products = pd.read_csv("data/products.csv")

customers.to_sql("customers", "sqlite:///retail.db", if_exists = "replace", index = False)
orders.to_sql("orders", "sqlite:///retail.db", if_exists = "replace", index = False)
products.to_sql("products", "sqlite:///retail.db", if_exists = "replace", index = False)

%load_ext sql
%config SqlMagic.style = "_DEPRECATED_DEFAULT"
%sql sqlite:///retail.db
```

# Explore the Data

## Task 1: Show all customers

```python
%%sql
SELECT * FROM customers
```

## Task 2: Show all products

```python
%%sql
SELECT * FROM products
```

## Task 3: Show all orders

```python
%%sql
SELECT * FROM orders
```

# Basic Analysis (SELECT + WHERE)

## Task 4: Find Gold members

```python
%%sql
SELECT * FROM customers WHERE Membership = "Gold"
```

## Task 5: Find Electronics products

```python
%%sql
SELECT * FROM products WHERE Category = "Electronics"
```

## Task 6: Find pending orders

```python
%%sql
SELECT * FROM orders WHERE Status = "Pending"
```

# Sorting Data

## Task 7: Sort products by price (highest first)

```python
%%sql
SELECT * FROM products ORDER BY Price DESC
```

## Task 8: Sort customers by join date (newest first)

```python
%%sql
SELECT * FROM customers ORDER BY JoinDate DESC
```

# LIMIT + OFFSET

## Task 9: Show first 3 customers

```python
%%sql
SELECT * FROM customers LIMIT 3
```

## Task 10: Skip first 2 orders and show next 3

```python
%%sql
SELECT * FROM orders LIMIT 3 OFFSET 2
```

# Combined Logic

## Task 11: Find Cairo customers sorted by age (youngest first)

```python
%%sql
SELECT * FROM customers WHERE City = "Cairo"
ORDER BY Age ASC
```

## Task 12: Find most expensive products (sort by price descending, limit 3)

```python
%%sql
SELECT * FROM products ORDER BY Price DESC LIMIT 3
```

## Task 13: Find customers with membership = 'Gold' sorted by age

```python
%%sql
SELECT * FROM customers WHERE Membership = "Gold"
ORDER BY Age DESC
```
