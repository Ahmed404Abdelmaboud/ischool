```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

```python
study = pd.read_csv("data/Daily study habits dataset.csv")
study
```

```python
fig, ax = plt.subplots(1, 2, sharex = True)
fig.set_size_inches(15, 5)
fig.suptitle("Focus level and study time accross the week")
sns.lineplot(data = study, x = "Day", y = "Focus Level (1–5)", ax = ax[0], marker = ".", markersize = 15)
plt.ylabel("Focus Level")
sns.lineplot(data = study, x = "Day", y = "Study Time (hours)", ax = ax[1], marker = ".", markersize = 15)
plt.ylabel("Study Time");
```

## Insights:
- أقل يوم من حيث مستوى التركيز وعدد ساعات المذاكرة هو السبت
- أعلى يوم من حيث مستوى التركيز وعدد ساعات المذاكرة هو الأربعاء
- أقل الأيام في الأغلب هي أيام الأجازات

```python
sns.scatterplot(data = study, x = "Focus Level (1–5)", y = "Study Time (hours)")
sns.regplot(data = study, x = "Focus Level (1–5)", y = "Study Time (hours)")
plt.title("Relation between focus level and study level")
plt.xlabel("Focus Level")
plt.ylabel("Study Time");
```
