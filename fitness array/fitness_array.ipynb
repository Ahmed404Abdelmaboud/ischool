```python
import numpy as np
```

```python
fitness = np.array([
    [15, 160, 50],
    [16, 165, 55],
    [17, 170, 60],
    [18, 175, 65],
    [19, 180, 70]
])
```

```python
# الأعمار
print(np.mean(fitness[ : , 0]))
# الأطوال
print(np.mean(fitness[ : , 1]))
# الأوزان
print(np.mean(fitness[ : , 2]))
```

```python
# الأعمار
print(np.std(fitness[ : , 0]))
# الأطوال
print(np.std(fitness[ : , 1]))
# الأوزان
print(np.std(fitness[ : , 2]))
```

```python
# الأعمار
print(np.min(fitness[ : , 0]))
# الأطوال
print(np.min(fitness[ : , 1]))
# الأوزان
print(np.min(fitness[ : , 2]))
```

```python
# الأعمار
print(np.max(fitness[ : , 0]))
# الأطوال
print(np.max(fitness[ : , 1]))
# الأوزان
print(np.max(fitness[ : , 2]))
```
