```python
# Data (student scores)
scores = [55, 80, 62, 95, 71, 74, 82, 60, 91, 200]
```

```python
scores = sorted(scores)
n = len(scores)
```

```python
scores
```

```python
def get_median(lst):
    m = len(lst) // 2
    if len(lst) % 2 == 0:
        return (lst[m-1] + lst[m]) / 2
    return lst[m]
```

```python
# min & max
min_value = scores[0]
max_value = scores[-1]
```

```python
# median
median = get_median(scores)
```

```python
# split data
mid = n // 2
lower = scores[ : mid]
upper = scores[mid : ] if n % 2 == 0 else scores[mid+1 : ]
```

```python
Q1 = get_median(lower)
Q3 = get_median(upper)
```

```python
IQR = Q3 - Q1
```

```python
print("Stored data : ", scores)
print("Min : ", min_value)
print("Q1 : ", Q1)
print("Median : ", median)
print("Q3 : ", Q3)
print("Max : ", max_value)
print("IQR : ", IQR)
```
